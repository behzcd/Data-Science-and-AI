## 8-VAZIFA. Toshkent shahri uchun, har bir yilda o'rtacha min va max haroratlarni ko'rsatuvchi grafik chizing. **Grafikni faylga saqlab oling**

```python
# JAVOBNI SHUYERGA YOZING
max = df['TM']
min = df['Tm']

plt.figure(figsize=(16,9))
# max grafigi
plt.plot(max,
         linewidth=1.5,
         linestyle='--',
         color='#8E44AD',
         alpha=0.7,
         marker='o',
         markerfacecolor = 'g')

# min grafigi
plt.plot(min,
         linewidth=1.5,
         linestyle='-',
         color='c',
         alpha=0.7,
         marker='o',
         markerfacecolor = 'r')

# Grafik nomi:
plt.title("Toshkent shahar uchun yillik o'rtacha harorat")
# x-o'q nomi:
plt.xlabel('Yil')
plt.ylabel('Harorat')

# Har bir chiziq haqida ma'lumot:
plt.legend(["o'rtacha maximum harorat", "o'rtacha minimum harorat"], loc=2, fontsize='medium')
plt.grid(color='k', alpha=0.5, linestyle='-') # setkani o'zgartirish
plt.savefig('Tash_temp_MinMax.png')
plt.show()
```

```python
#@title To'g'ri javob
minT = df['Tm']
maxT = df['TM']
plt.figure(figsize=(16,8))
plt.plot(minT, linewidth=2, linestyle="--", marker='h', markerfacecolor='g', label="O'rtacha minimum harorat")
plt.plot(maxT, linewidth=2, linestyle="-", marker='v', markerfacecolor='r', label="O'rtacha maksimum harorat")
plt.grid()
plt.legend()
plt.title("Toshkent shahri uchun yillik o'rtacha min va max harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.savefig('tempMinMax.png')
plt.show()
```
